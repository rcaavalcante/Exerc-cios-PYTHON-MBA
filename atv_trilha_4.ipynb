## **Imports**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px 
```

###**Import base 1**

```python
diplomas_ufg =  pd.read_csv(r'/content/10-2022.csv', sep=",")
diplomas_ufg = diplomas_ufg.sort_values(by='data_de_ingresso_no_curso', ascending=True)
dados_ufg = diplomas_ufg[['nome_curso_superior', 'data_de_ingresso_no_curso', 'data_da_expedicao_do_diploma']]
dados_ufg = dados_ufg.rename(columns={'nome_curso_superior':'Graduação', 'data_de_ingresso_no_curso':'Data_Matrícula', 'data_da_expedicao_do_diploma':'Data_Formatura'})
```

###**Criando o Gráfico 1**

```python
frequencia_cursos = dados_ufg['Graduação'].value_counts()
df = pd.DataFrame(frequencia_cursos)

df.plot.bar(figsize = (12, 8), color='b')

plt.legend(title='Quantidade de formandos por curso')
```

###**Interpretação gráfico 1**

O gráfico acima, gerado a partir da biblioteca matplotlib.pyplot, mostra que o curso de Agronomia foi o curso que mais emitiu diplomas no mês de Outubro, representando mais de 10% da amostragem total (418 formandos), com 44 diplomas; 5 diplomas a mais que o curso de Medicina Veterinária, segundo colocado. Ao todo, 30 cursos formaram alunos no mês de Outubro de 2022, com cursos como Ciências Sociais, Relações Públicas e Letras: Português, formando 1 estudante apenas.

###**Import base 3**

```python
video_game =  pd.read_csv(r"/content/List of best-selling video games.csv", sep=",")
#display(video_game)
video_game = video_game[['Rank', 'Title', 'Sales', 'Publisher(s)[a]']]
video_game = video_game[:10]
video_game['Sales'][0] = '238,000,000'
video_game = video_game.rename(columns={'Rank':'RANQUE','Title' : 'NOME', 'Sales': 'VENDAS', 'Publisher(s)[a]': 'Empresa'})

# Substitui a vírgula por ponto na coluna 'VENDAS'
video_game['VENDAS'] = video_game['VENDAS'].str.replace(',', '.')
# Remove os pontos da coluna 'VENDAS'
video_game['VENDAS'] = video_game['VENDAS'].str.replace('.', '')

# Converte a coluna 'VENDAS' para float
video_game['VENDAS'] = video_game['VENDAS'].astype(float)

```

###**Criando gráfico 3**

```python
plt.figure(figsize=(12,12))
plt.subplot(1, 1, 1)
sns.barplot(x=video_game.VENDAS, y=video_game.NOME) 
plt.xlabel('Titulo do Jogo')
plt.ylabel('Faturamento em bilhoes U$')
plt.title('Lista dos 10 videogames mais vendidos de todos os tempos', weight='bold')

plt.show()
```

###**Interpretação gráfico 3**

O gráfico acima foi gerado a partir de dados disponibilizados pela plataforma Kaggle e demonstra os 10 jogos de videogame que mais geraram lucro, em vendas. A tabela contém a relação de 49 jogos; portanto, para a amostragem no gráfico não ficar tão grande, e a visualização ficar mais intuitiva, durante a fase de tratamento de dados, foi filtrado somente o top 10. O jogo 'Minecraft' demonstra o melhor lucro, com uma margem grande de diferença do segundo colocado 'GTA V'. 


###**Criando gráfico 4**

```python
frequencia = video_game['Empresa'].value_counts()
frequencia_games = pd.DataFrame(frequencia)

frequencia_games.plot(figsize = (10, 6), color='g', yticks=range(0,5,1))
plt.xticks(rotation='vertical')

plt.legend(title='Empresas que mais aparecem no top 10')
```

###**Interpretação gráfico 4**
